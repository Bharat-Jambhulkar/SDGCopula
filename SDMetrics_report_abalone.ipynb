```python
from sdmetrics.reports.single_table import QualityReport
import pandas as pd
```

```python
original = pd.read_csv('abalone.csv')
synthetic = pd.read_csv('abalone_syn.csv')
```

```python
original.dtypes
```

```python
import json

with open('metadata_abalone.json') as f:
    my_metadata_dict = json.load(f)
```

```python
# Inspect and adjust metadata if needed
print(f"Columns in original data: {original.columns.tolist()}")
print(f"Columns in synthetic data: {synthetic.columns.tolist()}")
print(f"Columns in metadata: {list(my_metadata_dict['columns'].keys())}")
```

```python
report = QualityReport()
report.generate(original, synthetic, my_metadata_dict)
```

```python
report.get_score()
```

```python
fig = report.get_visualization(property_name='Column Shapes')
fig.show()
```

```python
from sdmetrics.reports.single_table import DiagnosticReport
report = DiagnosticReport()
report.generate(original, synthetic, my_metadata_dict)
```

```python
report.get_details(property_name='Data Validity')
```

```python
from sdmetrics.visualization import get_column_plot

fig = get_column_plot(
    real_data=original ,
    synthetic_data=synthetic,
    column_name='Whole weight',
    plot_type='distplot'
)

fig.show()
```

```python
from sdmetrics.visualization import get_column_pair_plot

fig = get_column_pair_plot(
    real_data=original,
    synthetic_data=synthetic,
    column_names=['Whole weight', 'Shucked weight'],
    plot_type='scatter'

)

fig.show()
```
